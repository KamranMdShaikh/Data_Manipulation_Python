##### Convert column attributes to column (Indicator_Name)

```python
df = pd.DataFrame({'Product':['pant','Pant','Pen','Box','CS','CS','Pen','Pant','Pen'],'Country_Code':['FR','FR','FR','USA','USA','USA','BR','BR','BR'],'Indicator_Name':['GPD','Pop','birth','GPD','Pop','birth','GPD','Pop','birth'], 'AGE':[14,34,56, 25, 67, 68, 55, 8,99]})
df
```

```python
# Group by desired columns and aggregate 'AGE' using dictionary comprehension
grouped_df = df.groupby(['Product', 'Country_Code']).apply(lambda x: dict(zip(x['Indicator_Name'], x['AGE']))).reset_index()

# Convert the dictionary columns to separate columns in the DataFrame
pivot_df = pd.DataFrame(grouped_df[0].tolist(), index=grouped_df.index)

# Concatenate the original columns with the new columns
pivot_df = pd.concat([grouped_df[['Product', 'Country_Code']], pivot_df], axis=1)

pivot_df
```

##### Convert column to column attributies (q1	q2	q3)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
'location': ['AL', 'TX', 'MO'],
'area': ['JN', 'AU', 'JC'],
'q1': [44, 25, 36],
'q2': [29, 32, 39],
'q3': [34, 43, 24],})

df
```

```python
df.melt(id_vars=["location", "area"], 
        var_name="quarter", 
        value_name="sales").sort_values(['location', 'quarter']).reset_index(drop=True)
```
